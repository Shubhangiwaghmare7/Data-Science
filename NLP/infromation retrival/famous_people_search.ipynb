```python
import numpy as np
import pandas as pd
import string
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer #TFid Vectorization
from sklearn.neighbors import KDTree
```

```python
person=pd.read_csv('famous_people.csv')
person
```

```python
#preprocessing
sw=stopwords.words('english')
```

```python
#tfid vectorized
tfid=TfidfVectorizer()
train_tfid=tfid.fit_transform(person.Text).toarray()
train_tfid
```

```python
#adding the vector to the table
person['tfid']=list(train_tfid)
person.head()
```

```python
#training on KDtree
kdtree=KDTree(train_tfid)
```

```python
#to do query on kd tree
distance,idx= kdtree.query(person.tfid[2].reshape(1,-1),k=3)

for i, value in list(enumerate(idx[0])):
    print(f'Name:     {person["Name"][value]}')
    print(f'distance: {distance[0][i]}')
    print(f'URI:      {person["URI"][value]}')
    
```
